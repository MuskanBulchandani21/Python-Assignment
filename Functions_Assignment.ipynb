#5.What are iterators in Python and how do they differ from iterables?
-In Python:

Iterable: Any object that can be looped over (like for loop) is called an iterable. Examples include lists, tuples, strings, etc.

Iterator: An iterator is an object that remembers where it is during iteration and returns items one at a time using the __next__() method. It is created from an iterable using the iter() function.

```python
#Example:
# Iterable
numbers = [1, 2, 3]  # This is an iterable

# Iterator
num_iter = iter(numbers)  # Creating an iterator from the iterable

print(next(num_iter))  # Output: 1
print(next(num_iter))  # Output: 2
```

#6.Explain the concept of generators in Python and how they are defined.
- What are Generators?
Generators are a special type of iterator in Python used to generate values on the fly without storing them in memory. They are useful when working with large datasets or infinite sequences.

 How are they defined?
Generators are defined using functions with the yield keyword instead of return. Each time the generator function is called using next(), it resumes from where it left off and yields the next value.

 Key Points:
Uses yield instead of return

Saves memory (no need to store all values at once)

Automatically implements the iterator protocol

```python
#Example:
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

# Using the generator
counter = count_up_to(3)

print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
print(next(counter))  # Output: 3
```

#7.What are the advantages of using generators over regular functions?
-Advantages of Generators:
Memory Efficient:
Generators don’t store all values in memory. They generate values on the fly, which is ideal for large datasets.

Faster Execution (for large data):
Since they yield one item at a time, they are faster than functions that return huge lists.

Simpler Code:
Generators simplify code using yield instead of writing custom iterator classes.

Pause and Resume:
With yield, a generator remembers its state, allowing you to pause and continue from where it left off.

```python
#Exampl:
#Regular Function (returns a full list):
def get_numbers(n):
    result = []
    for i in range(1, n+1):
        result.append(i)
    return result

print(get_numbers(5))

#Generator (yields values one-by-one):
def generate_numbers(n):
    for i in range(1, n+1):
        yield i

gen = generate_numbers(5)
for num in gen:
    print(num)
```

#8.What is a lambda function in Python and when is it typically used?
-A lambda function in Python is a small, anonymous function defined using the lambda keyword.
It can take any number of arguments but has only one expression

When is it Used?
Lambda functions are typically used when:

You need a quick, short function for a single use

Used with functions like map(), filter(), and sorted()

You don’t want to define a full function using def.


```python
#Example:
students = [("Alice", 85), ("Bob", 72), ("Charlie", 90)]

# Sort students by their score (second item in the tuple)
sorted_students = sorted(students, key=lambda x: x[1])

print(sorted_students)
```

#9.Explain the purpose and usage of the `map()` function in Python.
- Purpose:
The map() function in Python is used to apply a specific function to each item in an iterable (like a list or tuple) and returns a new iterable (a map object) containing the results.

- Usage:
It simplifies code by removing the need for a loop.

Commonly used with lambda functions for short, one-time operations.

The result from map() needs to be converted using list(), tuple(), etc., if you want to view the results directly.

```python
#Example:
# Convert a list of temperatures from Celsius to Fahrenheit
celsius = [0, 10, 20, 30]

# Formula: (C × 9/5) + 32 = F
fahrenheit = list(map(lambda x: (x * 9/5) + 32, celsius))

print(fahrenheit)
```

#10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- 1. map()
Purpose: Applies a function to every element in an iterable and returns a new iterable with the results.

- 2. filter()
Purpose: Filters items in an iterable based on a condition. Only the elements that return True for the function are kept.

- 3. reduce() (comes from functools module)
Purpose: Repeatedly applies a function to elements in the iterable and reduces them to a single result.

```python
#Example:
from functools import reduce

nums = [1, 2, 3, 4, 5]

# map: square each number
print(list(map(lambda x: x**2, nums)))      # [1, 4, 9, 16, 25]

# filter: keep even numbers
print(list(filter(lambda x: x % 2 == 0, nums)))  # [2, 4]

# reduce: multiply all numbers
print(reduce(lambda x, y: x * y, nums))     # 120
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
# Function to check if a number is prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# List all prime numbers from 1 to 200
prime_list = []
for i in range(1, 201):
    if is_prime(i):
        prime_list.append(i)

# Print the result
print("Prime numbers from 1 to 200 are:")
print(prime_list)
```

```python
#5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            next_val = self.a + self.b
            self.a, self.b = self.b, next_val
            self.count += 1
            return next_val

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

```python
#6.Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in power_of_two(5):
    print(power)
```

```python
#7.Implement a generator function that reads a file line by line and yields each line as a string.
# Step 1: Create a sample text file for demonstration
with open('sample.txt', 'w') as file:
    file.write("This is the first line\n")
    file.write("This is the second line\n")
    file.write("This is the third line\n")

# Step 2: Define the generator function
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # strip() removes \n from the end

# Step 3: Use the generator and print output
print("Reading file using generator:\n")
for line in read_file_line_by_line('sample.txt'):
    print(line)
```
